```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("housing.csv", header=None, sep="\s+")
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.head(2)
```

```python
# Establecer el estilo
sns.set_theme(style='whitegrid', context='notebook')
```

## Entendiendo los datos

Ya una vez descargados los datos lo que hay que hacer es entenderlos. Es importante saber que el algoritmo de regresi칩n lineal se basa altamente en correlaciones por lo que se recomienda que comencemos visualizando un gr치fico de correlaciones usando *pairplot* de Seaborn

```python
cols = ['DIS', 'INDUS', 'CRIM', 'RM', 'MEDV']
sns.pairplot(
    data= df[cols],
    height=2.5
)

plt.show()
```

Ya una vez visto c칩mo est치n correlacionadas las variables de manera visual es importante usar un m칠todo num칠rico para poder entender las correlaciones.

```python
cm = np.corrcoef(df[cols].values.T)
sns.set_theme(font_scale=1.5)
sns.heatmap(
    cm,
    cbar= True,
    annot= True,
    yticklabels= cols,
    xticklabels= cols,
)
```

En este mapa de calor podemos afirmar que el coeficiente de correlaci칩n lineal de las variables 'RM' y 'MEDV' es de 0.7, lo cual la hace una correlaci칩n positiva y altamente 칰til para crear nuestro modelo de regresi칩n lineal con Scikit-Learn
